```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
res = classifier("Today is the best day of my life!")

print(res)
```

```python
res = classifier("I miss the way things used to be.")
print(res)
```

```python
res = classifier("I don't know why, but I just broke down in tears.")
print(res)
```
